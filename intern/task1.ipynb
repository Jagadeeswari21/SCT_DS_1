```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'C:\Users\kala\Desktop\intern\API_SP.POP.TOTL_DS2_en_csv_v2_20333.csv', skiprows=4)
df.head() 
```

```python
print(df.columns)
```

```python
aggregates = ['World', 'High income', 'Low & middle income', 'Middle income',
              'Low income', 'Upper middle income', 'Lower middle income',
              'IDA & IBRD total', 'IBRD only', 'IDA total', 'East Asia & Pacific',
              'Europe & Central Asia', 'Latin America & Caribbean',
              'Sub-Saharan Africa', 'South Asia', 'North America', 
              'Early-demographic dividend', 'Late-demographic dividend', 
              'Sub-Saharan Africa (IDA & IBRD countries)', 'European Union', 
              'Arab World', 'OECD members']

df_2022 = df[~df['Country Name'].isin(aggregates)][['Country Name', '2022']].dropna()
df_2022 = df_2022.sort_values(by='2022', ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(y='Country Name', x='2022', data=df_2022, color='orange')
plt.title('Top 10 Most Populated Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```
